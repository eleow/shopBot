```python
glossary=pd.read_excel('/Users/shashanknigam/downloads/nlp_project/shopBot/webscraping/glossary.xls',index_col=0)
#Getting the doc representation of the each of the glossary keys. This is done with respect to en_core_web_md dataset.
doc_list = []
for i in glossary["Term"]:
    doc_list.append(nlp1(i))
doc_list = np.array(doc_list)
features = []
words_not_in_glosssary= []
non_nouns = ['twilight','horizon','lightning','performance','dos','lite','butter','legend','edge','flex','room','primacy','flow','spectrum','hbs','sirius','mk','edition','noise','solidege','armour','coach','fusion','pace','metal','peak','prestige','hx','omega','ace','mm','mini','piano','forte','twin','wasp','orb','oog','bullet','spark','infinity','fe','heaven','answer',"blasterx",'nano','iem','mkii','jones','signature','series','carbon','beoplay','home','street','pro','generation','smoke','air','conduction','ie','dash']

import re
pattern=re.compile('^[a-z]+$')
numbers = re.compile('[0-9]+')
numerics = ['single','double','triple','quad']
adj = ["sweatproof","limited"]
no_adj = ['ludacris','prime',"original",'remote','back','peak','elite','pulse','armour','skyline','active','free']
verb_listing = ['cancelling','isolating']
adv = ["superbass"]


scores = lambda x,doc1:np.array([doc1.similarity(x1) for x1 in x])

def matching_word(token,confidence=0.65):
    if token.vector_norm!=0:
        score = scores(doc_list,token)
        max_score_ind = np.argmax(score)
        max_score = np.max(score)
        if max_score>confidence:
            #print(max_score,doc_list[max_score_ind],token.text)
            return token.text
        else:
            #print("No Match",max_score,doc_list[max_score_ind],token.text)
            return ""
    
    else:
        return ""
```

```python
def getFeatureName(productName,Org):
    #print(Org)
    
    feature = re.compile('[a-z]+-[a-z]+ ')
    feature2 = re.compile('([0-9]{1,2}h playtime|[0-9]{1,2}hrs playtime|[0-9]{1,2} hours play[ ]*time|[0-9]{1,2} hour playtime|[0-9]{1,2}-hour playtime|[0-9]{1,2} hours|[0-9]{1,2} hour|[0-9]{1,2}h battery|[0-9]{1,2} hour battery)')
    
    ipx = re.compile('ipx[0-9]')
    length = re.compile('([0-9]{1,2}\.[0-9]{0,2}ft|[0-9]{1,2}\.[0-9]{0,2}m)')
    bluetooth = re.compile('bluetooth [0-9]\.[0-9]')
    compatible = re.compile('compatible with \w+')
    weight = re.compile('[0-9]{2}g')
    battery= re.compile('[0-9]{1,10}mah')
    drivers = re.compile('(dual [0-9]{2} mm drivers|[0-9]{2}mm driver)')
    jack = re.compile('[0-9]\.[0-9]mm jack')
    cvc = re.compile('cvc [0-9]\.[0-9]')
    suitable = re.compile('suitable for \w')
    soundsurge = re.compile('soundsurge [0-9]{2}')
    productName = productName.lower()
    productName = productName.replace(",","")
    productName = productName.replace(Org+" ","")
    productName = productName.replace(Org.replace(' ','-')+" ","")
    if Org == "bang & olufsen":
        productName = productName.replace('b&o play'+" ","")
    productName = productName.replace("]","")
    productName = productName.replace("&","")
    productName = productName.replace("[","")
    productName = productName.replace("/"," ")
    productName = productName.replace("\\","")
    productName = productName.replace("new","")
    productName = productName.replace("_","")
    productName = productName.replace("'","")
    productName = productName.replace("【","")
    productName = productName.replace("】","")
    productName = productName.replace("(","")
    productName = productName.replace(")","")
    productName = productName.replace("canceling","cancelling")
    productName = productName.replace("ture","true")
    
    productName=re.sub("[ ]+"," ",productName)
    feature_list = ['upgrade','hifi',"bass driven",'fits all','true wireless','figertip','folding','ipod','ipad','hanging','magnet controlled','metal finish','advanced sound effect',"charging box",'led digital display','gym','drive sport','sports','hands-free calling','smartphone','pistion fit','resistant to sweat','high resolution','mp3','mp4',"5.0","mini case","smallest",'lightest','best','rechargeable',"earphones",'magnetic connection','wireless','xbox one','ps4','nintendo switch','ios','android','running performance','running','one size','fast  charging','certified sound','deep bass','matte finish','microphone','headset','mobile','remote','sound deep bass','wired','dynamic','alexa enabled','cell phone','durable','lightweight','over ear','earbuds','mic','portable charging gyro box',"headphones",'ipx7 waterproof','ipx7','active noise cancelling','enabled with ar','bluetooth','active noise cancelling','active noise cancellation','noise reduction','noise isolation','ipx5','iphone','ip68']
    norog = ['kids','wired','in-ear','balanced','pistion','high','hi-fi','hi-res','best','coloud','matte','on-ear','mobile','active','bluetooth','over','gaming','heavy','foldable','noise','sleep','upgrade','ear','open','3.5mm','bone','new','bulk','bluetooth','in','bluetooth','earbud','wireless','usb','noise','earbuds','headphones','headphones','wholesale','true','earphones']
    prod_features = []
    prod_name = productName
    if len(prod_name.split())>1:
        if prod_name.split()[0] not in norog:
            prod= (" ".join(prod_name.split()[1:])).strip()
        else:
            prod =prod_name
        #print(prod)
        #Pos tagging for getting the features out.  
        if len(feature.findall(prod))>0:
            for j in feature.findall(prod):
                #print(j)
                prod_features.append(j.strip())
                prod=prod.replace(j.strip(),"")
                productName=productName.replace(j.strip(),"")
                productName=re.sub("[   ]+"," ",productName)
        if len(ipx.findall(prod))>0:
            for j in ipx.findall(prod):
                #print(j)
                prod_features.append(j.strip())
                prod=prod.replace(j.strip(),"")
                productName=productName.replace(j.strip(),"")
                productName=re.sub("[   ]+"," ",productName) 
        if len(battery.findall(prod))>0:
            for j in battery.findall(prod):
                #print(j)
                prod_features.append(j.strip())
                prod=prod.replace(j.strip(),"")
                productName=productName.replace(j.strip(),"")
                productName=re.sub("[   ]+"," ",productName)         
        if len(soundsurge.findall(prod))>0:
            for j in soundsurge.findall(prod):
                #print(j)
                prod_features.append(j.strip())
                prod=prod.replace(j.strip(),"")
                productName=productName.replace(j.strip(),"")
                productName=re.sub("[   ]+"," ",productName)
        if len(compatible.findall(prod))>0:
            for j in compatible.findall(prod):
                #print(j)
                prod_features.append(j.strip())
                prod=prod.replace(j.strip(),"")
                productName=productName.replace(j.strip(),"")
                productName=re.sub("[   ]+"," ",productName)        
        if len(weight.findall(prod))>0:
            for j in weight.findall(prod):
                #print(j)
                prod_features.append(j.strip())
                prod=prod.replace(j.strip(),"")
                productName=productName.replace(j.strip(),"")
                productName=re.sub("[   ]+"," ",productName)        
        if len(jack.findall(prod))>0:
            for j in jack.findall(prod):
                #print(j)
                prod_features.append(j.strip())
                prod=prod.replace(j.strip(),"")
                productName=productName.replace(j.strip(),"")
                productName=re.sub("[   ]+"," ",productName)
        if len(feature2.findall(prod))>0:
            for j in feature2.findall(prod):
                prod_features.append(j.strip())
                prod=prod.replace(j.strip(),"")
                productName=productName.replace(j.strip(),"")
                productName=re.sub("[   ]+"," ",productName)
        if len(bluetooth.findall(prod))>0:
            for j in bluetooth.findall(prod):
                prod_features.append(j.strip())
                prod=prod.replace(j.strip(),"")
                productName=productName.replace(j.strip(),"")
                productName=re.sub("[   ]+"," ",productName)
        if len(cvc.findall(prod))>0:
            for j in cvc.findall(prod):
                prod_features.append(j.strip())
                prod=prod.replace(j.strip(),"")
                productName=productName.replace(j.strip(),"") 
                productName=re.sub("[   ]+"," ",productName)
        if len(drivers.findall(prod))>0:
            for j in drivers.findall(prod):
                prod_features.append(j.strip())
                prod=prod.replace(j.strip(),"")
                productName=productName.replace(j.strip(),"")
                productName=re.sub("[   ]+"," ",productName)
        if len(suitable.findall(prod))>0:
            for j in suitable.findall(prod):
                prod_features.append(j.strip())
                prod=prod.replace(j.strip(),"")
                productName=productName.replace(j.strip(),"")
                productName=re.sub("[   ]+"," ",productName)
        for i in feature_list:
            #print(i)
            if i in prod:
                prod_features.append(i)
                prod=prod.replace(i,"")
                productName=productName.replace(i,"")
                productName=re.sub("[   ]+"," ",productName)
        #print(prod)
        #print(productName)
        doc=nlp1(prod)
        feat = ""
        for token in doc:
            #print(token,token.pos_)
            #Only alphabetic: issue
            if pattern.match(token.text):
                #print(token.text)
                
                if len(token.text)==1:
                    pass
                
                #print(token.text,token.pos_)
                elif token.pos_ in ('NOUN','ADJ','ADV') and token.text not in non_nouns:
                    if (token.text in numerics) and doc[token.i+1].text=="driver":
                        feat+=token.text+" "
                        #prod_features.append(token.text+" "+doc[token.i+1].text)
                        #productName=productName.replace(token.text+" "+doc[token.i+1].text,"")
                    else:
                        if token.pos_ in ('ADJ','ADV'):
                            if feat!="":
                                feat +=token.text+" " 
                                feat_doc=nlp1(feat.strip())
                                wm = matching_word(feat_doc)
                                if wm!="":
                                    prod_features.append(wm)
                                    productName=productName.replace(wm,"")
                                    feat=""
                                #Indicating if the token has a match but not the entire document 
                                else:
                                    wm = matching_word(token)
                                    if wm!="":
                                        prod_features.append(wm)
                                        productName=productName.replace(wm,"")
                                        #feat does not map here thus adding the feat to  non existing glossary
                                        prod_features.append(feat)
                                        productName=productName.replace(feat,"")
                                        words_not_in_glosssary.append(feat)
                                        feat=""
                            elif token.text in adj:
                                prod_features.append(token.text)
                                productName=productName.replace(token.text,"")
                                words_not_in_glosssary.append(token.text)
                            elif token.text in no_adj:
                                pass
                            else:
                                wm = matching_word(token)
                                if wm!="":
                                    prod_features.append(wm)
                                    productName=productName.replace(wm,"")
                                else:
                                    feat+=token.text+" "
                        else: #Word considered is a noun
                            if feat!="": 
                                feat +=token.text+" "
                                feat_doc=nlp1(feat.strip())
                                wm = matching_word(feat_doc)
                                if wm!="":
                                    prod_features.append(wm)
                                    productName=productName.replace(wm,"")
                                    feat=""
                                #Indicating if the token has a match but not the entire document 
                                else:
                                    wm = matching_word(token)
                                    if wm!="":
                                        prod_features.append(wm)
                                        productName=productName.replace(wm,"")
                                        #feat does not map here thus adding the feat to  non existing glossary
                                        feat=feat.replace(token.text,"").strip()
                                        prod_features.append(feat)
                                        productName=productName.replace(feat,"")
                                        words_not_in_glosssary.append(feat)
                                        feat=""
                                    else:
                                        prod_features.append(feat.strip())
                                        productName=productName.replace(feat.strip(),"")
                                        words_not_in_glosssary.append(feat.strip())
                                        feat=""
                            else:
                                if token.text == "bone" and (doc[token.i+1].text=="conductor" or doc[token.i+1].text=="conduction"): 
                                    words_not_in_glosssary.append(token.text+" "+doc[token.i+1].text)
                                    prod_features.append(token.text+" "+doc[token.i+1].text)
                                    productName=productName.replace(token.text+" "+doc[token.i+1].text,"")
                                else:
                                    wm = matching_word(token)
                                    if wm=="":
                                        words_not_in_glosssary.append(token.text)
                                    prod_features.append(token.text)
                                    productName=productName.replace(token.text,"")
                elif (token.pos_ =="ADP" or token.pos_=="PART") and (feat=="" or token.text=="in") and matching_word(token,confidence=0.6):
                    #print(token.text)
                    if token.text in ["in",'on']:
                        if doc[token.i+1].text=="ear":
                            feat+=token.text+" "

                        
                    else:
                        feat+=token.text+" "
                    #print(feat)
                elif (token.pos_ =="ADP" or token.pos_=="PART")  and feat=="":
                    #print(token.text)
                    productName=productName.replace(token.text,"")  
                elif token.pos_=="DET" and feat!="":
                    feat+=token.text+" "
                elif token.pos_ == "VERB":
                    if token.text in verb_listing:
                        feat+=doc[token.i-1].text+" "+token.text
                        wm = matching_word(nlp1(feat))
                        if wm=="":
                            words_not_in_glosssary.append(feat)
                        prod_features.append(feat)
                        productName=productName.replace(feat,"")
                        feat=""
                    elif token.text in colors:
                        prod_features.append(token.text)
                        productName = productName.replace(token.text,"")
                    elif (token.text =="closed" or token.text=="open") or (token.i<len(doc)-1 and doc[token.i+1].text)=="back":
                        if(token.i<len(doc)-1):
                            prod_features.append(token.text+" "+doc[token.i+1].text)
                            words_not_in_glosssary.append(token.text+" "+doc[token.i+1].text)
                            productName=productName.replace(token.text+" "+doc[token.i+1].text,"")
                elif token.pos_=="ADV" and token.text in adv:
                    words_not_in_glosssary.append(token.text)
                    prod_features.append(token.text)
                    productName=productName.replace(token.text,"") 
                elif token.pos_=="CCONJ" and token.text=="and":
                      productName=productName.replace(token.text,"")   
                elif token.text in colors:
                    prod_features.append(token.text)
                    productName = productName.replace(token.text,"")        
            elif token.pos_ =="NUM" and numbers.match(token.text) and token.i<(len(doc)-1) and doc[token.i+1].text=="driver":    
                feat+=token.text+" "
            
        if feat!="":
            prod_features.append(feat)
            productName= productName.replace(feat.strip(),"")
            words_not_in_glosssary.append(feat)
            feat=""
    prod_features = ",".join(prod_features)
    for i in productName.split(' '):
        if len(i)==1:
            productName=productName.replace(" "+i,'')
    productName=re.sub('works','',productName)
    
    productName=re.sub('，','',productName)
    productName=re.sub('-','',productName)
    productName=re.sub('–','',productName)
    productName=re.sub("[   ]+"," ",productName)
    
    #productName=re.sub(" ","",productName)
    #productName=productName.replace("  "," ")
    return (productName.strip(),prod_features)
```
